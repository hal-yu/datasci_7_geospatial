## Geospatial Data Processing and Visualization

### Dataset 1: National Obesity by State

```python
df1 = gpd.read_file('National_Obesity_By_State.geojson')
df1
```

```python
df1.plot('Obesity', legend=True)
```

This dataset contains information about the national obesity percentages in the United States based on the state. From the map, you can see the range on the right, from yellow to violet. The darker the colors, the lower the obesity rate. From this chart, you can see that the highest obesity rate is next to Texas, which is Louisiana and the lowest obesity rate is Colorado.

### Dataset 2: Cancer Rates

```python
df2 = gpd.read_file('Cancer_Rates.geojson')
df2
```

```python
df2.plot('All_Cancer', legend=True)
```

This dataset contains information about the cancer rates of the county Lake Country in Illinois. From the legend on the right, you can see that the zip codes with higher cancer rates are in yellow, and the zip codes with lower cancer rates are violet. From this, we can see that the the one in yellow, zip 60069 has the highest cancer rates and the two in violet, with zip codes 60085 and 60073 have the lowest cancer rates.
